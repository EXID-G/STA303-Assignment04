```python
#内嵌画图，省略掉plt.show()这一步，直接显示图像。
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# plt.rcParams['font.sans-serif'] = ['SimHei']  #防止中文乱码，还可以有Microsoft YaHei
# plt.rcParams["axes.unicode_minus"] = False   #字符显示，防止负号乱码

import warnings
warnings.filterwarnings('ignore')
```

```python
plt.style.use("ggplot")
```

## load data

```python
resnet18_fashionminist_data = pd.read_csv('./csvdata/resnet18_fashionminist.CSV')
densenet121_fashionminist_data = pd.read_csv('./csvdata/densenet121_fashionminist.CSV')

resnet18_cifar100_data = pd.read_csv('./csvdata/resnet18_cifar100.CSV')
densenet121_cifar100_data = pd.read_csv('./csvdata/densenet121_cifar100.CSV')
```

```python
plt.figure(figsize=(20, 12))
sns.set_context(font_scale=4)
plt.suptitle("The scatter plot of coverage_rate and average_size",fontsize = 24)

plt.subplot(221) #构建两行两列，并在第一个子图绘图
sns.scatterplot(x = "coverage_rate", y = "average_size",hue="predictor",style="score_function",s = 250, data = resnet18_fashionminist_data,alpha = 1);
plt.title("Scatter plot of resnet18_fashionminist",fontsize = 18)

plt.subplot(222) #构建两行两列，并在第2个子图绘图
sns.scatterplot(x = "coverage_rate", y = "average_size",hue="predictor",style="score_function",s = 250, data = densenet121_fashionminist_data,alpha = 1);
plt.title("Scatter plot of densenet121_fashionminist",fontsize = 18)

plt.subplot(223) #构建两行两列，并在第3个子图绘图
sns.scatterplot(x = "coverage_rate", y = "average_size",hue="predictor",style="score_function",s = 250, data = resnet18_cifar100_data,alpha = 1);
plt.title("Scatter plot of resnet18_cifar100",fontsize = 18)

plt.subplot(224) #构建两行两列，并在第4个子图绘图
sns.scatterplot(x = "coverage_rate", y = "average_size",hue="predictor",style="score_function",s = 250, data = densenet121_cifar100_data,alpha = 1);
plt.title("Scatter plot of densenet121_cifar100",fontsize = 18)


plt.tight_layout()
plt.savefig("./figure/Scatter plot.svg")
```
